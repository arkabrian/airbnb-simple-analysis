<h1>Final Project AirBNB</h1>

<img src='airbnb.png'>

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("airbnb.csv")
data
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
print("Rata - rata     :", data['price'].mean())
print("Median          :", data['price'].median())
print("Standar Deviasi :", data['price'].std())
```

```python
data['room_type'].value_counts()
```

<h2>Entire Home/Apt</h2>

<img src='apartment.jpg'>

```python
data_home = data[data['room_type'] == 'Entire home/apt']
data_home
```

```python
data_home.isna().sum()
```

```python
data_home['name'].fillna(data_home['id'], inplace = True)
data_home.isna().sum()
```

```python
data_home['host_name'].fillna(data_home['host_id'], inplace= True)
data_home.isna().sum()
```

```python
data_home['reviews_per_month'].fillna(data_home['reviews_per_month'].mean(), inplace = True)
data_home.isna().sum()
```

```python
data_home['last_review'].fillna(data_home['last_review'].mode()[0], inplace = True)
data_home.isna().sum()
```

```python
data_home_subset = data_home[['name', 'neighbourhood_group', 'price', 'minimum_nights', 'number_of_reviews', 'last_review','reviews_per_month','availability_365']]
data_home_subset
```

```python
data_home_subset.mean()
```

```python
data_home_subset.median()
```

```python
data_home_subset.std()
```

```python
data_home['neighbourhood_group'].value_counts()
```

```python
data_home_mean = data_home.groupby(['neighbourhood_group']).mean()
plt.bar(data_home_mean.index, data_home_mean['price'] ,label='mean', width=0.25)
```

```python
data_home_median = data_home.groupby(['neighbourhood_group']).median()
plt.bar(data_home_median.index, data_home_median['price'] ,label='median', width=0.25)
```

```python
data_home_std = data_home.groupby(['neighbourhood_group']).std()
plt.bar(data_home_std.index, data_home_std['price'] ,label='median', width=0.25)
```

```python
x = np.arange(len(data_home_mean.index))
width = 0.25

plt.bar(x-width, data_home_mean['price'] ,label='mean', width=width)
plt.bar(data_home_median.index, data_home_median['price'] ,label='median', width=width)
plt.bar(x+width, data_home_std['price'] ,label='std', width=width)

plt.title('Average price in every neighbourhood')

plt.tight_layout()
plt.legend()

plt.show()
```

<h3>Setelah tau rata - ratanya, kita sekarang bisa nentuin klasifikasi harganya</h3>

```python
data_home_subset['Price Class'] = np.where(data_home_subset['price'] > 200, 'Expensive', 
                                 np.where((data_home_subset['price'] <= 200) & (data_home_subset['price'] > 100), 'Medium', 'Cheap'))
data_home_subset.head(10)
```

```python
data_home_subset['Price Class'].value_counts().plot.bar()
```

<h3>Data 2019</h3>

```python
data_home_subset['year'] = pd.DatetimeIndex(data_home_subset['last_review']).year
data_home_2019 = data_home_subset[data_home_subset['year'] > 2018]
data_home_2019
```

```python
data_home_2019['Price Class'].value_counts().plot.bar()
```

```python
data_home_2019.nsmallest(10,['price'])
```

```python
data_home_2019.nlargest(10,['price'])
```

```python
print("Harga termurah di 2019 adalah",data_home_2019['price'].min())
print("Harga termahal di 2019 adalah",data_home_2019['price'].max())
```
